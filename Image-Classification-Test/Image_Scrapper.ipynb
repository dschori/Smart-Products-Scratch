```python
from icrawler.builtin import GoogleImageCrawler, BingImageCrawler
import os

IMAGE_DIR = 'Download'
QUERYS = ['cat', 'dog', 'cow', 'sheep']
```

```python
for query in QUERYS:
    google_crawler = BingImageCrawler(feeder_threads=2,
                                        parser_threads=4,
                                        downloader_threads=8,
                                        storage={'root_dir': '{}/{}'.format(IMAGE_DIR, query)})
    try:
        idx_offset = len(os.listdir('{}/{}'.format(IMAGE_DIR, query)))
    except FileNotFoundError:
        idx_offset = 0
        
    google_crawler.crawl(keyword=query, max_num=700, file_idx_offset=idx_offset)
```

## Create Directorys for Training Data and Count Images

```python
from pathlib import Path

for data_split in ['training', 'validation', 'testing']:
    for query in QUERYS:
        Path("./Image_Data/{}/{}".format(data_split, query)).mkdir(parents=True, exist_ok=True)
```

```python
file_lengths = {}
for query in QUERYS:
    file_lengths[query] = len(os.listdir('./{}/{}'.format(IMAGE_DIR, query)))
print('Pictures per Class: {}'.format(file_lengths))
```

```python
MIN_IMAGES = min(file_lengths.values())
print('Class "{}" has the least pictures: {}'.format(min(file_lengths, key=file_lengths.get), MIN_IMAGES))
```

## Process Images and store them

```python
import skimage
import numpy as np
from skimage.io import imread
from skimage.io import imsave
from skimage import img_as_float, img_as_ubyte
from skimage.color import rgba2rgb, gray2rgb
from skimage.transform import resize
from sklearn.model_selection import train_test_split
import threading

CLASSES = QUERYS
IMAGE_HEIGTH = 224
IMAGE_WIDTH = 224

def process_images_multithreading(images, image_path, save_path, c):
    train_images, val_images = train_test_split(images, train_size=0.8, random_state=1)
    val_images, test_images = train_test_split(val_images, train_size=0.6, random_state=1)
    
    while True:
        print('Processing {} Train Images for Class "{}"...'.format(len(train_images), c))
        process_images(image_list=train_images, image_path='./{}/{}'.format(IMAGE_DIR, c), save_path='./Image_Data/training/{}'.format(c))
        print('Processing {} Validation Images for Class "{}"...'.format(len(val_images), c))
        process_images(image_list=val_images, image_path='./{}/{}'.format(IMAGE_DIR, c), save_path='./Image_Data/validation/{}'.format(c))
        print('Processing {} Test Images for Class "{}"...'.format(len(test_images), c))
        process_images(image_list=test_images, image_path='./{}/{}'.format(IMAGE_DIR, c), save_path='./Image_Data/testing/{}'.format(c))
        
        return
    
def process_images(image_list, image_path, save_path):
    for image_name in image_list:
        img = load_image('{}/{}'.format(image_path, image_name))
        img = resize_image(img, heigth=IMAGE_HEIGTH, width=IMAGE_WIDTH)
        image_name, image_extension = os.path.splitext(image_name)
        save_image(img, save_path=save_path, image_name=image_name)

def load_image(path):
    img = imread(path)
    img = img_as_float(img)
    if img.shape[-1] == 4:
        img = rgba2rgb(img)
    if len(img.shape) == 2:
        img = gray2rgb(img)
    return img

def resize_image(img, heigth, width):
    img = resize(img, (heigth, width))
    return img

def save_image(img, save_path, image_name):
    img = np.clip(img, -1.0, 1.0)
    img = img_as_ubyte(img)
    imsave('./{}/{}.jpg'.format(save_path, image_name), img)
    
threads = []

for c in CLASSES:
    images = os.listdir('./{}/{}'.format(IMAGE_DIR, c))
    images = images[:MIN_IMAGES]
    
    t = threading.Thread(target=process_images_multithreading, args=(images, './{}/{}'.format(IMAGE_DIR, c), './Image_Data/training/{}'.format(c), c, ))
    threads.append(t)
    t.start()
    
for t in threads:
    t.join()
    
print('Finished!')
```

## Display Sample Images

```python
import matplotlib.pyplot as plt
import random

def show_images(split, n=4):
    rows, columns = n, len(CLASSES)

    fig, axis = plt.subplots(rows, columns, figsize=(columns*5, rows*3))

    for c, cl in enumerate(CLASSES):
        images = os.listdir('./Image_Data/{}/{}/'.format(split, cl))
        random.shuffle(images)
        images = images[:rows]
        for j in range(rows):
            img = load_image(path='./Image_Data/{}/{}/{}'.format(split, cl, images[j]))
            axis[j, c].imshow(img)
            axis[j, c].axis('off')
            axis[j, c].set_title(cl)

    plt.tight_layout()
    plt.show()
```

### Train Images:

```python
show_images(split='training', n=4)
```

### Validation Images:

```python
show_images(split='validation', n=3)
```

### Test Images:

```python
show_images(split='testing', n=2)
```
